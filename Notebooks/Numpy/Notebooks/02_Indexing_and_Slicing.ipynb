```python
import pandas as pd
import numpy as np
```

```python
sales = pd.read_csv("../../../Data/sales.csv")
```

```python

# Q1: Access individual elements of a NumPy array using positive and negative indexing.
```

```python
quantity = sales['quantity'].to_numpy()
```

```python
print(quantity[0])
print(quantity[1])
print(quantity[-1])
print(quantity[-2])
```

```python
# Q2: Extract multiple elements from a NumPy array using slicing.
```

```python
# first 5 elements
print(quantity[:5])

# last 5 elements
print(quantity[-5:])

# between index 10 and 20
print(quantity[10:21])
```

```python
# Q3: Reverse the order of elements in a NumPy array.
```

```python
# Full reverse
print(quantity[::-1])

# Reverse and extract the last three elements
print(quantity[-1:-4:-1])
```

```python
# Q4: Access rows and columns from a 2D NumPy array using indexing.
```

```python
matrix = sales.to_numpy()
matrix.shape
```

```python

# Q4: 2D Array Slicing
# Practice slicing on the 2D array `matrix` to access specific rows and columns.
```

```python


# Q4.1 Display the first 5 rows and all columns.
print(matrix[:5, :])

# Q4.2 Display all rows and the first 3 columns.
print(matrix[:, :3])

# Q4.3 Display the first 10 rows and the first 2 columns.
print(matrix[:10, :2])

# Q4.4 Display the last 5 rows and the last 2 columns.
print(matrix[-5:, -2:])

# Q4.5 Display rows 10 to 19 and columns 1 to 2.
print(matrix[10:20, 1:3])

# Q4.6 Display every second row and all columns.
print(matrix[::2, :])

# Q4.7 Display all rows with columns in reverse order.
print(matrix[:, ::-1])

# Q4.8 Display the first 5 rows with columns in reverse order.
print(matrix[:5, ::-1]) 

# Q4.9 Display the last 10 rows and the first 2 columns.
print(matrix[-10  :, :2])

# Q4.10 Display every second row and every second column.
print(matrix[::2, ::2])
```
