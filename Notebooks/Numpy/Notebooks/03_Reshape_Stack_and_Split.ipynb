```python
import pandas as pd
import numpy as np
```

```python
sales = pd.read_csv("../../../Data/sales.csv")
quantity = sales["quantity"].to_numpy()
matrix = sales.to_numpy()
```

```python
# Q.1 Take the first 100 Quantity values and reshape them into a 10 × 10 matrix.
```

```python
quantity_matrix = quantity[:100].reshape(10, 10)
```

```python
print(quantity_matrix)
```

```python
# Q.2 Flatten the reshaped matrix back into a 1D array.
```

```python
quantity_flattened = quantity_matrix.flatten()
print(quantity_flattened)
```

```python
# Q3: Horizontally stack Product ID and Quantity
```

```python
product_2D = sales["product_id"].to_numpy().reshape(-1,1)
quantity_2D = sales["quantity"].to_numpy().reshape(-1,1)
```

```python
stacked_matrix =np.hstack((product_2D, quantity_2D))
print(stacked_matrix)
```

```python
# Q4: Vertically stack two equal-sized Quantity arrays
```

```python
q1 = quantity[:50]
q2 = quantity[50:100]
```

```python
stacked_quantity = np.vstack((q1, q2))
print(stacked_quantity.shape)
```
